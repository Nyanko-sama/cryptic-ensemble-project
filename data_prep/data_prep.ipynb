## Pick 20 APO structures for initial testing

Just so I don't generate 1000 proteins in a wrong way

```python
train_data = load_dataset(f'{CRYPTOBENCH_PATH}/cryptobench-dataset/folds/train-fold-0.json')
apo_chains = extract_apo_chains(train_data)

# create json with apo pdb ids where the chain doesn't contain a "-" and the same format as the original train_data.json
filtered_apo_chains = {apo_pdb_id: chain for apo_pdb_id, chain in apo_chains.items() if "-" not in chain}
# keep only first 20 apo pdb ids 
filtered_apo_chains = dict(list(filtered_apo_chains.items())[:20])
filtered_train_data = {apo_pdb_id: train_data[apo_pdb_id] for apo_pdb_id in filtered_apo_chains.keys()}

with open(f'20_train_structures.json', 'w') as f:
    json.dump(filtered_train_data, f, indent=4)
```
